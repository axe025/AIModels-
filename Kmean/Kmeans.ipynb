# **Unsupervised clustering algorithm**

##**Select a dataset**

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/13_kmeans/income.csv")
df.head()
```

```python
plt.scatter(df["Age"], df["Income($)"])
```

## **Implement a K-means**

```python
km=KMeans(n_clusters=3,n_init=5, random_state=100)
km
```

```python
y_Predict=km.fit_predict(df[['Age','Income($)']])
y_Predict
```

```python
df["cluster"]=y_Predict
df
```

### **Visualize the results**

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1["Age"],df1["Income($)"],color="green")
plt.scatter(df2["Age"],df2["Income($)"],color="red")
plt.scatter(df3["Age"],df3["Income($)"],color="black")

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="purple",marker="*",label="centroid")
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()
```

```python
#Here we see onre group are clustered well but the centroids of other groups are not right and its due our scaling on Xaixs are pretty narrow
# so the centroids are not right
```

## **Scaling numeric features**

```python
scl=MinMaxScaler()
scl.fit(df[['Age','Income($)']])
df[['Age','Income($)']]=scl.transform(df[['Age','Income($)']])
df
```

### **Implement a K-means**

```python
km=KMeans(n_clusters=3,n_init=5, random_state=100)
km
```

```python
y_Predict=km.fit_predict(df[['Age','Income($)']])
y_Predict
```

```python
df["cluster"]=y_Predict
df.head()
```

### **Visualize the results**

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1["Age"],df1["Income($)"],color="green")
plt.scatter(df2["Age"],df2["Income($)"],color="red")
plt.scatter(df3["Age"],df3["Income($)"],color="black")

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="purple",marker="*",label="centroid")
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()
```
